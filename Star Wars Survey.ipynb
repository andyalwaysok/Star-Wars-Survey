## Cleaning columns 15 through 28, which contain data on the characters respondents view favorably and unfavorably.

- Which character do respondents like the most?
- Which character do respondents dislike the most?
- Which character is the most controversial (split between likes and dislikes)?

```python
df.rename({'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.': 'Han Solo',
          'Unnamed: 16': 'Luke Skywalker', 'Unnamed: 17':'Princess Leia Organa', 'Unnamed: 18': 'Anakin Skywalker',
          'Unnamed: 19': 'Obi Wan Kenobi', 'Unnamed: 20': 'Emperor Palpatine', 'Unnamed: 21': 'Darth Vader', 'Unnamed: 22': 'Lando Calrissian',
          'Unnamed: 23': 'Boba Fett', 'Unnamed: 24': 'C-3P0', 'Unnamed: 25': 'R2 D2',
          'Unnamed: 26': 'Jar Jar Binks', 'Unnamed: 27': 'Padme Amidala', 'Unnamed: 28': 'Yoda'}, axis = 1, inplace = True)
```

```python
df[df.columns[17]].value_counts()
```

```python
def favor_function(val):
    if val == 'Very favorably':
        return 2
    elif val == 'Somewhat favorably':
        return 1
    elif val == 'Neither favorably nor unfavorably (neutral)' or val == 'Unfamiliar (N/A)' or val == np.nan:
        return 0
    elif val == 'Somewhat unfavorably':
        return -1
    elif val == 'Very unfavorably':
        return -2
    else:
        return 0

df[df.columns[15:29]] = df[df.columns[15:29]].applymap(favor_function)
df[df.columns[15:29]]
```

```python
df[df.columns[15:29]].sum().sort_values()
```

```python
characters = df[df.columns[15:29]].sum().sort_values().plot(kind='barh', alpha=0.9, color = '#af0b1e', figsize=(8,5), legend=False)
characters.title.set_text('Star Wars Chacters respondents view favorably and unfavorably')
for location in ['left','right','top','bottom']:
    characters.spines[location].set_visible(False)
characters.xaxis.tick_top()    
characters.tick_params(top=False, left=False)
```

- Respondants like Han Solo the most
- Rspondants dislike Jar Jar Binks the most
- It looks like Anakin Skywalker, Lando Calrissian, Padme Amidala, Darth Vader, Boba Fett are the most controversial ones
